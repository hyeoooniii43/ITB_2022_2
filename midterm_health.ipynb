```python
%matplotlib inline
import pandas as pd # pandas 로딩

# data 가져오기
df = pd.read_csv('https://raw.githubusercontent.com/hyeoooniii43/ITB_2022_2/main/health_data.csv')

display(df.head(10))   # 처음 10행 확인
```

```python
df.shape    # 전체 행과 열의 개수 출력
df.info()     # 10000개의 행 중에서 결측지가 있는 행과 타입 확인 가능.
```

```python
# 사용할 컬럼명 바꾸기
df = df.rename (columns = {'연령대코드(5세단위)': '연령대코드'})
df = df.rename (columns = {'신장(5Cm단위)': '신장'})
display(df.head(10))
```

```python
#1. 성별 신장에 대한 시각화
import seaborn as sns # seaborn 로딩

sns.boxplot(data = df, x = '성별코드', y = '신장')
```

<결과 해석>

*   남성의 신장 평균이 여성의 신장 평균보다 높음.
*   남성의 신장은 극단적으로 높거나 낮은 키가 많이 보이고, 쫍은 범위에 분포하며 수염이 짧음.
*   여성의 신장은 중앙값을 중심으로 넓은 범위에 분포하고 있고, 하나의 낮은 키가 있음.







```python
#2. 나이별 시력 차이 알아보기

df_eye = df.copy()

# 이상치 확인
df_eye.describe() # 시력의 max 값이 9.9인 이상치 발생
```

```python
# 이상치 제거
df_eye2 = df_eye[df_eye['시력(좌)'] < 9.9]
df_eye3 = df_eye2[df_eye2['시력(우)'] < 9.9]

df_eye3.describe() # 9.9 값 모두 제거 완료
```

```python
sns.barplot(data = df_eye3, x = '연령대코드', y = '시력(좌)')
```

```python
sns.barplot(data = df_eye3, x = '연령대코드', y = '시력(우)')
```

<결과 해석>
1.   점점 나이가 들수록 시력이 낮아짐.
2.   좌우 시력의 평균은 거의 비슷하게 나타남.

```python
#3. 흡연상태 한눈에 알아보기

df_1 = df.copy()
# 결측치 확인
display(df_1['흡연상태'].isna().sum()) # 흡연상태에서 2가지 있는 거 확인
```

```python
# 결측지 제거
df1 = df_1.dropna(subset = ['흡연상태'])
display(df1['흡연상태'].isna().sum()) # 결측치 제거 완료
```

```python
# 파이 차트로 시각화하기 위한 빈도수 계산하여 새로운 Data frame 만들기
df2 = df1.groupby('흡연상태' , as_index = False).agg( count = ('흡연상태', 'count'))
df2
```

```python
# 파이 차트 만들기
plt.figure(figsize=(5,5))
plt.pie(df2['count'], labels= df2['흡연상태'], autopct='%.1f%%')
plt.show()
```

<결과 해석>
*   현재 비흡연이라고 말한 사람들의 비율이 62.3%로 압도적으로 높음.
*   다음으로 예전에 피웠지만 현재는 피지 않는 사람들의 비율이 19.3%로 나타나고 있고, 흡연자와 흡연 경험이 있는 사람들은 전체의 37.7%임.
